## Pregunta 1

Primero, creamos el DataFrame para la Pregunta 1, haciendo un merge con los datos del DataFrame "codigos" junto con el de "poblacion". Luego, agrupamos los datos obtenidos por Región, con el criterio de Población y los sumamos para obtener el valor total de población en cada región del país.

````python
df_preg1 = df_codigos.merge(df_poblacion, left_on='Código Comuna 2018', right_on='CODIGO')
df_preg1 = df_preg1['POBLACIÓN'].groupby([df_preg1['Nombre Región']]).sum()
df_preg1
````

Por último, generamos un Gráfico de Barras para visualizar los datos de una mejor manera. Cabe destacar que la cantidad de población por región fluctua normalmente, exceptuando en la Región Metropolitana de Santiago, la cual posee una amplia diferencia en su cantitad de  habitantes.

````python
fig1 = plt.figure(figsize=(30,14))
plt.title("Población por Región")
plt.bar(df_preg1.keys(),df_preg1)
plt.xticks(rotation=90)
````

## Pregunta 2

En primer lugar, creamos los DataFrame correspondientes a la Región de Valparaíso y Metropolitana, que contienen el nombre de la comuna y su respectivo presupuesto. Cabe destacar que, para mejorar la visualización de los datos de la Región de Valparaíso, eliminamos la comuna de Algarrobo, ya que esta correspondía a un outlier.

````python
df_preg2 = df_codigos.merge(df_presupuesto_2019, left_on='Código Comuna 2018', right_on='CODIGO')
valparaiso = df_preg2[(df_preg2["Nombre Región"]=="Valparaíso")]
valparaiso = valparaiso[['Nombre Comuna','PRESUPUESTO']]
valparaiso = valparaiso.drop(63)
valparaiso
````

````python
metropolitana = df_preg2[(df_preg2["Nombre Región"]=="Metropolitana de Santiago")]
metropolitana = metropolitana[['Nombre Comuna','PRESUPUESTO']]
metropolitana
````

A continuación, realizamos una transformación a escala logarítmica a los presupuestos de las comunas de ambas regiones, para facilitar la visualización de los datos.

````python
valparaiso_log= valparaiso.copy()
valparaiso_log['PRESUPUESTO'] = np.log(valparaiso_log['PRESUPUESTO'] + 1)

metropolitana_log= metropolitana.copy()
metropolitana_log['PRESUPUESTO'] = np.log(metropolitana_log['PRESUPUESTO'] + 1)
````

Ahora, para simplificar la comprensión de los datos de cada región, generamos sus respectivos Histogramas.

````python
sns.histplot(data=valparaiso_log, x="PRESUPUESTO",  color="blue", alpha=0.5,bins=10)
plt.title('Histograma Presupuesto Comunas de la Región de Valparaíso');
````

````python
sns.histplot(data=metropolitana_log, x="PRESUPUESTO",  color="red", alpha=0.5,bins=10)
plt.title('Histograma Presupuesto Comunas de la Región Metropolitana');
````

Por último, generamos un Histograma Múltiple, el cual nos permite comparar los datos obtenidos de ambas regiones. Podemos notar una gran diferencia entre las comunas de la Región Metropolitana y la de Valparaíso, en donde para la Región Metropolitana, la cantidad de comunas con un ***menor*** presupuesto, se asimilan a la cantidad de comunas con un ***mayor*** presupuesto correspondientes a la Región de Valparaíso. Además, es posible destacar la gran cantidad de comunas dentro de la Región Metropolitana que poseen un elevado presupuesto.

````python
sns.set_theme()
sns.histplot(data=valparaiso_log, x="PRESUPUESTO",  color="blue", alpha=0.5,bins=10, label="Valparaíso")
sns.histplot(data=metropolitana_log, x="PRESUPUESTO",  color="red", alpha=0.5,bins=10, label="Metropolitana")
plt.title('Histograma Presupuesto Comunas de la Región de Valparaíso y Metropolitana');
plt.legend(loc="upper left")
````

## Pregunta 3

Para comenzar, generamos el DataFrame correspondiente a la Pregunta 3, el cual contiene las comunas, su presupuesto, población, código, su respectiva región y también si es o no parte de la Región Metropolitana.

````python
df_preg3 = df_poblacion.merge(df_presupuesto_2019[['CODIGO', 'PRESUPUESTO']], left_on='CODIGO', right_on='CODIGO')
df_preg3 = df_preg3.merge(df_codigos, left_on='CODIGO', right_on='Código Comuna 2018')
df_preg3 = df_preg3[['CODIGO', 'Nombre Región', 'Nombre Comuna', 'PRESUPUESTO', 'POBLACIÓN' ]]
df_preg3.loc[df_preg3['Nombre Región'] != 'Metropolitana de Santiago', 'Capital'] = 'Otra Región'
df_preg3.loc[df_preg3['Nombre Región'] == 'Metropolitana de Santiago', 'Capital'] = 'Metropolitana' 
df_preg3
````

A continuación, realizamos una transformación a escala logarítmica a los presupuestos y población de cada comuna.

````python
df_preg3['PRESUPUESTO'] = np.log(df_preg3['PRESUPUESTO'] + 1)

df_preg3['POBLACIÓN'] = np.log(df_preg3['POBLACIÓN'] + 1)
````

Por último, realizamos un Gráfico de Dispersión de estos parámetros, para las comunas, diferenciando aquellas que pertenecen a la Región Metropolitana. Al ver este gráfico, es lógico concluir que a mayor población, habrá un mayor presupuesto comunal. Además, cabe destacar que aquellas comunas pertenecientes a la Región Metropolitana tienen una mayor población y por ende, un mayor presupuesto, exceptuando algunos casos específicos. Esto permite ver que gran parte de la población Chilena y el presupuesto comunal está alojado en esta región.

````python

sns.scatterplot(x="POBLACIÓN", y="PRESUPUESTO", data=df_preg3, hue="Capital")
````
