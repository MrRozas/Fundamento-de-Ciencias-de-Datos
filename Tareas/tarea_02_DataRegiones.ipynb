````python
#Juntamos los datos necesarios, y dejamos solo los datos de Valparaíso
region_psu = pd.merge(psu, codigos[['Código Comuna 2018','Nombre Región']], left_on='CODIGO',  right_on = 'Código Comuna 2018')
psu_valparaiso = region_psu.loc[region_psu.loc[:, 'Nombre Región'] == 'Valparaíso']

#Eliminamos las comunas que no importen
psu_valparaiso.drop('CODIGO', axis=1, inplace=True) 
psu_valparaiso.drop('Código Comuna 2018', axis=1, inplace=True)
psu_valparaiso.drop('Nombre Región', axis=1, inplace=True)

#Renombramos las columnas dado que los números generan conflicto con el código
psu_valparaiso = psu_valparaiso.rename(columns={'450_MUNICIPALES':'MUNICIPALES','450_PARTICULARES':'PARTICULARES', '450_SUBVENCIONADOS':'SUBVENCIONADOS'}) #Código compartido por mi pareja de tarea

#Eliminamos los datos no recepcionados
no_recep=psu_valparaiso.query('MUNICIPALES_PRIMERA == "No Recepcionado"').index.tolist()
psu_valparaiso=psu_valparaiso.drop(no_recep, axis=0)

no_recep=psu_valparaiso.query('PARTICULARES_PRIMERA == "No Recepcionado"').index.tolist()
psu_valparaiso=psu_valparaiso.drop(no_recep, axis=0)

no_recep=psu_valparaiso.query('SUBVENCIONADOS_PRIMERA == "No Recepcionado"').index.tolist()
psu_valparaiso=psu_valparaiso.drop(no_recep, axis=0)

no_recep=psu_valparaiso.query('MUNICIPALES == "No Recepcionado"').index.tolist()
psu_valparaiso=psu_valparaiso.drop(no_recep, axis=0)

no_recep=psu_valparaiso.query('PARTICULARES == "No Recepcionado"').index.tolist()
psu_valparaiso=psu_valparaiso.drop(no_recep, axis=0)

no_recep=psu_valparaiso.query('SUBVENCIONADOS == "No Recepcionado"').index.tolist()
psu_valparaiso=psu_valparaiso.drop(no_recep, axis=0)

#########################################

#Ahora convertimos los datos de object a valor númerico
psu_valparaiso['MUNICIPALES_PRIMERA'].convert_dtypes()
psu_valparaiso.MUNICIPALES_PRIMERA = pd.to_numeric(psu_valparaiso.MUNICIPALES_PRIMERA, errors='coerce')

psu_valparaiso['PARTICULARES_PRIMERA'].convert_dtypes()
psu_valparaiso.PARTICULARES_PRIMERA = pd.to_numeric(psu_valparaiso.PARTICULARES_PRIMERA, errors='coerce')

psu_valparaiso['SUBVENCIONADOS_PRIMERA'].convert_dtypes()
psu_valparaiso.SUBVENCIONADOS_PRIMERA = pd.to_numeric(psu_valparaiso.SUBVENCIONADOS_PRIMERA, errors='coerce')

psu_valparaiso['MUNICIPALES'].convert_dtypes()
psu_valparaiso.MUNICIPALES = pd.to_numeric(psu_valparaiso.MUNICIPALES, errors='coerce')

psu_valparaiso['PARTICULARES'].convert_dtypes()
psu_valparaiso.PARTICULARES = pd.to_numeric(psu_valparaiso.PARTICULARES, errors='coerce')

psu_valparaiso['SUBVENCIONADOS'].convert_dtypes()
psu_valparaiso.SUBVENCIONADOS = pd.to_numeric(psu_valparaiso.SUBVENCIONADOS, errors='coerce')


#Se eliminan los valores NA resultantes de las proporciones pedidas
prop_municipales = psu_valparaiso['MUNICIPALES']/psu_valparaiso['MUNICIPALES_PRIMERA']
prop_municipales = prop_municipales.dropna()
prop_municipales = prop_municipales.to_frame()
prop_municipales.columns=['Proporción Municipales']

prop_particulares = psu_valparaiso['PARTICULARES']/psu_valparaiso['PARTICULARES_PRIMERA']
prop_particulares = prop_particulares.dropna()
prop_particulares = prop_particulares.to_frame()
prop_particulares.columns=['Proporción Particulares']

prop_subvencionados = psu_valparaiso['SUBVENCIONADOS']/psu_valparaiso['SUBVENCIONADOS_PRIMERA']
prop_subvencionados = prop_subvencionados.to_frame()
prop_subvencionados = prop_subvencionados.dropna()
prop_subvencionados.columns=['Proporción Subvecionados']

#Respuestas/Analisis
print("Del gráfico generado, podemos ver los colegios particulares son los que presentan una proporción menor a los otros 2 tipos de colegio, con lo que se puede determinar que la enseñanza que aquí se imparte es capaz de llegar de mmanera distribuida a sus alumnos, y además de mejor manera, ya que la proporción de alumnos que saco más de 450 del total va mucho más de la manos que los otros 2 tipos, por lo que se puede ver una evidente desigualdad en la eduación, ya que se necesita dinero para tener buena educación.")
print("Respecto a los outliers, los 3 tipos de colegios los presentan, siendo los colegios subvencionados el con mayor, luego los municipales y por ultimo los particulares, apuntando en la misma dseiguladad antes mencionada, pero demostrando que hay alumnos que pueden escapar de este 'regla'.")
print("-------------------------------------------------------------")
print("Boxplots con la distribución de las proporciones de los distintos tipos de colegios de las comunas de Valparaiso, entre los que al darla por primera vez obtuvieron más de 450 en la PSU y los que la dieron por primera vez")
print("")
print("En este caso los tipos de colegio son: 0 = Municipales, 1 = Particulares, 2 = Subvencionados")
print("")

#Hacemos el boxplot
eje = sns.boxplot(data = [prop_municipales,prop_particulares,prop_subvencionados])
eje.set_title("Distribución proporciones Valparaiso", fontsize=20)
eje.set_ylabel("Proporción", fontsize=15)
eje.set_xlabel("Tipo de colegio", fontsize=15)
````
