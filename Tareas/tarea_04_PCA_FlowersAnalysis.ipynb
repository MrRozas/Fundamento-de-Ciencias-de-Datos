````python
from sklearn import datasets
import pandas as pd

iris = datasets.load_iris()
iris.keys()
df = pd.DataFrame(iris['data'])
df.columns = iris.feature_names
df
````

````python
# Recuerda que las flores Virgínica son las de tipo 2
iris['target']
````

````python
# Continúa la solución acá

#Primero dividimos en prueba y test
tipo_flores = pd.DataFrame(iris['target'])
X = pd.DataFrame(iris['data'], columns=iris['feature_names'])
X.insert(4, "Tipo de flores", tipo_flores, allow_duplicates=False)
X, y = X[['sepal length (cm)','sepal width (cm)', 'petal length (cm)','petal width (cm)' ]], X[['Tipo de flores']]
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3)
````

````python
#Estandarizamos el dataset de prueba
std_scaler = StandardScaler()
X_std = std_scaler.fit_transform(X_train)
X_std_test = std_scaler.fit_transform(X_test)

# Ahora vamos a reducir la dimensión del dataset estandarizado
# Queremos reducir a dos dimensiones
pca = PCA(n_components=2)
X_2D = pca.fit_transform(X_std)
X_2D = pd.DataFrame(X_2D, columns=["dimension 1", "dimension 2"])

X_2D_test = pca.fit_transform(X_std_test)
X_2D_test = pd.DataFrame(X_2D_test, columns=["dimension 1", "dimension 2"])

X_2D
````

````python
#Estandarizamos el dataset
std_scaler2 = StandardScaler()
X_std2 = std_scaler2.fit_transform(X_train)
X_std2_test = std_scaler2.fit_transform(X_test)

# Queremos reducir a una dimension
pca2 = PCA(n_components=1)
X_1D = pca2.fit_transform(X_std2)
X_1D = pd.DataFrame(X_1D, columns=["dimension 1"])

X_1D_test = pca2.fit_transform(X_std2_test)
X_1D_test = pd.DataFrame(X_1D_test, columns=["dimension 1"])

X_1D
````

````python
#Predecimos en 2D

#Modelo a entrenar
sgd_clf = SGDClassifier()
sgd_clf.fit(X_2D, y_train)

#Predecimos
y_pred = sgd_clf.predict(X_2D_test)
y_pred
````

````python
#Predecimos en 1D

#Modelo a entrenar
sgd_clf2 = SGDClassifier()
sgd_clf2.fit(X_1D, y_train)

#Predecimos
y_pred2 = sgd_clf2.predict(X_1D_test)
y_pred2
````

````python
#Predecimos para el dataset original

#Modelo a entrenar
sgd_clf3 = SGDClassifier()
sgd_clf3.fit(X_train, y_train)

#Predecimos
y_pred3 = sgd_clf3.predict(X_test)
y_pred3
````

````python
#Analisis para 2D

#Calculamos accuracy

print(accuracy_score(y_test, y_pred))

#Calculamos precision

print(precision_score(y_test, y_pred, average=None))

#Calculamos recall
print(recall_score(y_test, y_pred, average=None))

#Calculamos la varianza
print(pca.explained_variance_ratio_)
````

````python
#Analisis para 1D

#Calculamos accuracy

print(accuracy_score(y_test, y_pred2))

#Calculamos precision

print(precision_score(y_test, y_pred2, average=None))

#Calculamos recall
print(recall_score(y_test, y_pred2, average=None))

#Calculamos la varianza
print(pca2.explained_variance_ratio_)
````

````python
#Analisis para todo el dataset

#Calculamos accuracy

print(accuracy_score(y_test, y_pred3))

#Calculamos precision

print(precision_score(y_test, y_pred3, average=None))

#Calculamos recall
print(recall_score(y_test, y_pred3, average=None))

#Calculamos la varianza
#print(pca.explained_variance_ratio_)
````

En el caso donde se utiliza el dataset completo para el clasificador, vemos que de la muestra tomada, el 'accuracy' es de un 0.955, valor alto.
````python
#Hacemos la reducción con TSNE del dataset original, sin estandarizar
X = pd.DataFrame(iris['data'], columns=iris['feature_names'])
y = iris['target']
tsne = TSNE(n_components=2)
X_2D = tsne.fit_transform(X)
X_2D = pd.DataFrame(X_2D, columns=["dimension 1", "dimension 2"])
````

````python
#Luego lo graficamos

cmap_bold = ['#a72618', '#34892d', '#262ea7']

plt.figure(figsize=(5, 5))

plt.scatter(X_2D[y==0]["dimension 1"], X_2D[y==0]["dimension 2"], c=cmap_bold[0], label='Iris Setosa')
plt.scatter(X_2D[y==1]["dimension 1"], X_2D[y==1]["dimension 2"], c=cmap_bold[1], label='Iris Versicolor')
plt.scatter(X_2D[y==2]["dimension 1"], X_2D[y==2]["dimension 2"], c=cmap_bold[2], label='Iris Virginica')

plt.legend()

plt.show()
````

````python
#Ahora hacemos la reducción del dataset original con PCA

X = pd.DataFrame(iris['data'], columns=iris['feature_names'])
y = iris['target']

std_scaler = StandardScaler()
X_std = std_scaler.fit_transform(X)

# Queremos reducir a dos dimensiones
pca = PCA(n_components=2)

# Guardamos el dataset pero ahora en su versión reducida en 2 dimensiones
X_2D = pca.fit_transform(X_std)
X_2D = pd.DataFrame(X_2D, columns=["dimension 1", "dimension 2"])

cmap_bold = ['#a72618', '#34892d', '#262ea7']

plt.scatter(X_2D[y==0]["dimension 1"], X_2D[y==0]["dimension 2"], c=cmap_bold[0], label='Iris Setosa')
plt.scatter(X_2D[y==1]["dimension 1"], X_2D[y==1]["dimension 2"], c=cmap_bold[1], label='Iris Versicolor')
plt.scatter(X_2D[y==2]["dimension 1"], X_2D[y==2]["dimension 2"], c=cmap_bold[2], label='Iris Virginica')

plt.legend()
plt.show()
````

### Analisis:

En este caso, el primer gráfico corresponde a la reducción del dataset con TSNE a dos dimensiones. El segundo gráfico es lo mismo pero ahora con PCA.

La primera diferencia que notamos es que en el primero los datos están más concentrados, mientras que en el otro, si bien se ve una zona donde se mantienen los colores, están más distribuidos. Otra diferencia es la posición de los puntos, en el caso de TSNE, ubica los puntos de una manera totalmente distintas que en el caso de PCA.

Al ser un dataset pequeño, no es necesario la utilización de TSNE, que es más preciso, además de que complica el asunto. Por lo que PCA es lo mejor en este caso.

## Pregunta 3

````python
from sklearn.datasets import fetch_openml

mnist = fetch_openml('mnist_784', version=1, as_frame=True)
M, y = mnist['data'], mnist['target']
y = y.astype(np.uint8)

M, y = M[60000:], y[60000:]
````

````python
# Queremos reducir n dimensiones

#Código para poder determinar cuantos componentes son necesarios para que la varianza acumulada sea de 0.95
for i in range(2,785):
  n = i
  pca = PCA(n_components= n)
  M1 = pca.fit_transform(M)
  N1 = pca.explained_variance_ratio_
  #display(N1)
  a=0
  for j in range(0,n):
    a+=N1[j]
  if a>=0.95:
    break
n
````

````python
#Código para crear 2 listas con los valores de los componentes y sus respectivas varianzas acumuladas
lis=[]
lis2=[]
for i in range(2,155,5):
  n = i
  if i==152:
    n=150
  lis.append(n)
  pca = PCA(n_components= n)
  M2 = pca.fit_transform(M)
  N2 = pca.explained_variance_ratio_
  a = 0
  for j in range(0,n):
    a+=N1[j]
  lis2.append(a)
````

````python
#Graficamos
plt.plot(lis, lis2)
plt.title("Gráfico nro. de componentes vs Varianza acumulada", size=15)
plt.xlabel("Número de componentes")
plt.ylabel("Varianza acumulada")
plt.show()
````

### Analisis:
Se necesita una reducción de 150 elementos del dataset MNIST para poder obtener una varianza acumulada de 0.95. Si vamos tomando los valores de la varianza acumulada partiendo con n=2 hasta n=150, se obtiene el gráfico anterior. Aquí se ve que sigue una distribución logaritmica. Aquí se puede ver que a medida que aumentamos el número de componentes, tambien aumenta la varianza acumulada, pero no de manera lineal, sino que aumenta rapidamente al principio, y luego se va estancando en valores cercanos a 1.
