## Pregunta 2

````python
# Escriba su código aquí
#Genera y muestre un Dataframe que contenga solamente las siguientes columnas: Nombre Región, Nombre Provincia, Nombre Comuna, metros_plaza

df_metros_plaza = pd.merge(codigos, datos_comunas, left_on='Código Comuna 2018', right_on='id', how = 'left')

print("DataFrame con las columnas de: Nombre Región, Nombre Provincia, Nombre Comuna, metros_plaza")
print("")

df_metros_plaza = df_metros_plaza[['Nombre Región', "Nombre Provincia", "Nombre Comuna","metros_plaza"]]

display(df_metros_plaza)
````

## Pregunta 3

````python
# Escriba su código aquí
#¿Cuántos metros de plaza tiene CADA región?

sum_metros_region = df_metros_plaza.groupby(df_metros_plaza['Nombre Región'])['metros_plaza'].sum()

print("Metros de plaza por región:")
print("")

sum_metros_region
````

## Pregunta 4

````python
# Escriba su código aquí
#¿Cuál es la región con la mayor cantidad de metros de plaza?

df_sum_metros_region = sum_metros_region.to_frame()

print("Metros plaza por región ordenados de mayor a menor")
print("")

mayor_metros = df_sum_metros_region.sort_values(by='metros_plaza', ascending=False)
display(mayor_metros)

print("")

display(mayor_metros[:1])

print("")
print("La región con mayor cantidad de metros de plaza es la de Valparaíso.")
````

## Pregunta 5

````python
# Escriba su código aquí
#Genera un dataframe que contenga las siguientes columnas: Nombre comuna, Presupuesto, Población.

df_pres_pob = pd.merge(presupuesto, poblacion, on='CODIGO')

df_presupuesto_poblacion = df_pres_pob[['MUNICIPIO_x','PRESUPUESTO','POBLACIÓN']]

df_presupuesto_poblacion.columns = ['Nombre Comuna', 'Presupuesto','Población']

print("DataFrame con las columnas: Nombre comuna, Presupuesto, Población.")
print("")

df_presupuesto_poblacion
````

## Pregunta 6

````python
# Escriba su código aquí
#Entrega la población de cada región.

df_pob_reg = pd.merge(codigos, poblacion, left_on='Código Comuna 2018', right_on='CODIGO')
sum_poblacion_region = df_pob_reg.groupby(df_pob_reg['Nombre Región'])['POBLACIÓN'].sum()
print("Población por región:")
print("")
sum_poblacion_region
````

## Pregunta 7

````python
# Escriba su código aquí
#Entrega la población de cada provincia
print("Población por provincia:")
print("")
sum_poblacion_provincia = df_pob_reg.groupby(df_pob_reg['Nombre Provincia'])['POBLACIÓN'].sum()
sum_poblacion_provincia
````
