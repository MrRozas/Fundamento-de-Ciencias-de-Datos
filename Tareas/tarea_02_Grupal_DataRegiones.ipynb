## Pregunta 1

````python
merge = pd.merge(codigos, poblacion, left_on='Código Comuna 2018', right_on='CODIGO')
sum_poblacion= (merge.groupby(merge['Código Región'])[("POBLACIÓN")].sum())
# display(sum_poblacion)
plt.figure()
plt.title("Población por Región")
plt.xlabel("Número de Región")
plt.ylabel("Millones de personas")
plt.bar(sum_poblacion.keys(), sum_poblacion)

````

## Pregunta 2

````python
#Genere una visualización que muestre la distribución de los presupuestos comunales de la Región de Valparaiso junto con la distribución de presupuestos comunales de la Región Metropolitana. ¿Qué puede comentar de este gráfico?
pres_comu = pd.merge(codigos, presupuesto, left_on='Código Comuna 2018', right_on='CODIGO')
presupuesto_comunal = pres_comu[['Código Región','Nombre Región','Nombre Comuna','PRESUPUESTO']]
#display(presupuesto_comunal)

#Para solo dejar la region metropolitana, se identifican las filas que son de la region. 
metropolitana = presupuesto_comunal.loc[:, 'Código Región'] == 13
metropolitana = (presupuesto_comunal.loc[metropolitana])

valpo = presupuesto_comunal.loc[:, 'Código Región'] == 5
valpo = (presupuesto_comunal.loc[valpo])

pres_metropolitana_valpo = metropolitana.append(valpo)
pres_metropolitana_valpo['PRESUPUESTO']=pres_metropolitana_valpo['PRESUPUESTO'].astype('int')
pres_metropolitana_valpo['PRESUPUESTO'] = np.log(pres_metropolitana_valpo['PRESUPUESTO'] + 1)

#Histograma

sns.set_theme
sns.histplot(data=pres_metropolitana_valpo, x="PRESUPUESTO", hue="Nombre Región",bins=5, alpha=0.4)
plt.title('Histograma Presupuesto Región de Valparaíso y Metropolitana');

````

## Pregunta 3

````python
#Genere una visualización que muestre cómo se relaciona el presupuesto comunal y la población de las comunas en Chile. Además, incluya en el mismo gráfico, 
#la información correspondiente a si la comuna pertenece o no a la región metropolitana. ¿Qué puede comentar de la relación entre estas dos variables? 
#¿Se observan diferencias entre la región metropolitana vs otras regiones del país?

presupuesto['PRESUPUESTO'].convert_dtypes()
presupuesto.PRESUPUESTO = pd.to_numeric(presupuesto.PRESUPUESTO, errors='coerce')
presupuesto_poblacion = pd.merge(poblacion, presupuesto[['CODIGO','PRESUPUESTO']], on="CODIGO")

presupuesto_poblacion = pd.merge(presupuesto_poblacion, codigos[['Código Comuna 2018','Nombre Región']], left_on='CODIGO', right_on='Código Comuna 2018')
presupuesto_poblacion.drop('CODIGO', axis=1, inplace=True)
presupuesto_poblacion.drop('Código Comuna 2018', axis=1, inplace=True)

lista = ['Tarapacá', 'Antofagasta', 'Atacama','Coquimbo','Valparaíso', "Libertador General Bernardo O'Higgins", 'Maule','Biobío','La Araucanía', 'Los lagos', 'Aysén del General Carlos Ibáñez del Campo', 'Magallanes y de la Antártica Chilena', 'Metropolitana de Santiago', 'Los Ríos', 'Arica y Parinacota', 'Ñuble']

for i in lista:
  presupuesto_poblacion['Nombre Región'] = presupuesto_poblacion['Nombre Región'].str.replace(i, 'Otra')


presupuesto_poblacion['PRESUPUESTO'] = np.log(presupuesto_poblacion['PRESUPUESTO'] + 1)
presupuesto_poblacion['POBLACIÓN'] = np.log(presupuesto_poblacion['POBLACIÓN'] + 1)


my_plot = presupuesto_poblacion.plot("POBLACIÓN", "PRESUPUESTO", kind="scatter")
plt.show()
````
